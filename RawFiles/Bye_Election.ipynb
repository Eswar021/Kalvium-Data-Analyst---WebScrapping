```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.get('https://results.eci.gov.in/AcResultByeJune2024/index.htm')
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
boxes = soup.find_all('div', class_ = "col-md-3 col-12")
boxes
```

```python
dic = dict()
```

```python
data = []
for div in soup.find_all('div', class_='const-box'):
    link = div.find('a', href=True)['href']
    constituency = div.find('h3').text
    state = div.find('h4').text
    status = div.find('h2').text
    candidate = div.find('h5').text
    party = div.find('h6').text

    data.append({
        'Link': link,
        'Constituency': constituency,
        'State': state,
        'Status': status,
        'Candidate': candidate,
        'Party': party
    })

# Convert the list of dictionaries to a pandas DataFrame
Bye_Election = pd.DataFrame(data)

# # Save to CSV
# df.to_csv('election_results.csv', index=False)

# # Save to Excel
# df.to_excel('election_results.xlsx', index=False)
Bye_Election
```

```python
Bye_Election["Update_link"] = Bye_Election.Link.str.replace("candidateswise-", "https://results.eci.gov.in/AcResultByeJune2024/Constituencywise")
```

```python
update_links = Bye_Election["Update_link"].to_list()
update_links
```

```python
dfs = []

# Loop through each link and extract data
for link in Bye_Election["Update_link"]:
    response = requests.get(link)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract state name and remove parentheses
    state_name = soup.find('div', class_='page-title').find('h2').find('strong').text.strip('()')
    Constituency_name = soup.find('div', class_='page-title').find('h2').find('span').text.split('(')[0].strip()
    # Find the table in the HTML
    table = soup.find('table')
    
    # Extract headers
    headers = [th.text.strip() for th in table.find('thead').find_all('th')]
    
    # Extract rows
    rows = []
    for tr in table.find('tbody').find_all('tr'):
        cells = [td.text.strip() for td in tr.find_all('td')]
        rows.append(cells)
    
    # Create DataFrame for current link
    df = pd.DataFrame(rows, columns=headers)
    
    # Add state name as a new column without parentheses
    df['State'] = state_name
    df['Constituency'] = Constituency_name
    
    # Append current DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames
Bye_Election_Constituencywise = pd.concat(dfs, ignore_index=True)

# Display the final DataFrame
Bye_Election_Constituencywise.State = Bye_Election_Constituencywise.State.str.replace("(","")
Bye_Election_Constituencywise

# Optionally save the final DataFrame to a CSV or Excel file
# Bye_Election_Constituencywise.to_csv('Bye_Election_Constituencywise.csv', index=False)
# Bye_Election_Constituencywise.to_excel('Bye_Election_Constituencywise.xlsx', index=False)
```
